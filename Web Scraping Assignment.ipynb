```python
Links = []

start = 0
end = 67

for page in range(start, end):
    linkss_tag = WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.XPATH,'//a[@class="link-hover link text-x1 font-semibold"]')))
    
    for i in linkss_tag:
        Links.append(i.get_attribute('href'))
        
    next_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')))
    next_button.click()

    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//a[@class="link-hover link text-x1 font-semibold"]')))
    time.sleep(2) 

print(Links)
```
